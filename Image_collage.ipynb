```python
!pip install opencv-python
import numpy as np
import cv2
import pandas as pd
import matplotlib.pyplot as plt
```

```python
top_left = cv2.imread("C://Users/My PC/Downloads/top_left.jpg")
top_left = cv2.cvtColor(top_left,cv2.COLOR_BGR2RGB)
top_right = cv2.imread("C://Users/My PC/Downloads/top_right.jpg")
top_right = cv2.cvtColor(top_right,cv2.COLOR_BGR2RGB)
bottom_left = cv2.imread("C://Users/My PC/Downloads/bottom_left.jpg")
bottom_left = cv2.cvtColor(bottom_left,cv2.COLOR_BGR2RGB)
bottom_right = cv2.imread("C://Users/My PC/Downloads/bottom_right.jpg")
bottom_right = cv2.cvtColor(bottom_right,cv2.COLOR_BGR2RGB)
center = cv2.imread("C://Users/My PC/Downloads/center.jpeg")
center = cv2.cvtColor(center,cv2.COLOR_BGR2RGB)

top_left      = cv2.resize(top_left,(200,200))
top_right     = cv2.resize(top_right,(200,200))
bottom_left   = cv2.resize(bottom_left,(200,200))
bottom_right  = cv2.resize(bottom_right,(200,200))
center        = cv2.resize(center,(200,200))
```

```python
top_left      = cv2.copyMakeBorder(top_left,10,5,10,5,0)
top_right     = cv2.copyMakeBorder(top_right,10,5,5,10,0)
bottom_left   = cv2.copyMakeBorder(bottom_left,5,10,10,5,0)
bottom_right  = cv2.copyMakeBorder(bottom_right,5,10,5,10,0)
center        = cv2.copyMakeBorder(center,10,10,10,10,0)
```

```python
res1 = np.concatenate((top_left,top_right),axis = 1)
res2 = np.concatenate((bottom_left,bottom_right),axis = 1)
res =  np.concatenate((res1,res2),axis = 0)
res[110:330,110:330]=center
plt.imshow(res)
print(res.shape)
```
